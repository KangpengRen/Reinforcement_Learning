# Random Variable（随机变量）

随机变量是一个未知的量，其取值只取决于随机试验的结果。
- 通常使用大写字母`X`表示随机变量本身；
- 通常使用小写字母`x`表示随机变量的值。

# Probability Density Function(PDF)

- 通常使用$\mathcal{X}$表示随机变量定义域；
- 对于连续分布：
    $$
    \int_{\mathcal{X}} p(x) dx = 1.
    $$
- 对于离散分布：
    $$
    \sum_{x \in \mathcal{X}} p(x) = 1.
    $$

# Expectation

- 对于连续分布：
    $$
    \mathbb{E} = \int_{\mathcal{X}} p(x) f(x) dx.
    $$
- 对于离散分布：
    $$
    \mathbb{E} = \sum_{x \in \mathcal{X}} p(x) f(x).
    $$

# Random Sampling

一个箱子里有2个红球、5个绿球、3个篮球，随机从箱子里取出一个球，它是每种颜色的概率？

```python
from numpy.random import choice

samples = choice(["R", "G", "B"], size=100, p=[0.2, 0.5, 0.3])
print(samples)
```

重复100次，每取出一次记录颜色，其结果具有统计意义。

# Terminologies

`state`、`action`：**for example**，比如在游戏《超级马里奥》中，当前游戏的画面（马里奥的位置、板栗怪的位置、金币的位置……）即`状态state`，而当前马里奥可以采取的操作（如前进、后退、跳跃……）即`行为action`。其中，马里奥叫做`agent`。

`policy`($\pi$)：观测智能体的`state`，从而做出相应的`action`。
- 策略函数（本质为概率密度函数，强化学习就是在学习这个函数）：$\pi: (s, a) \rightarrow [0, 1]: \pi(a|s) = \mathbb{P} (A = a | S = s).$
- 反应在状态`s`时采取行为`a`的概率：
    - $\pi(left | s) = 0.2$,
    - $\pi(right | s) = 0.1$,
    - $\pi(up | s) = 0.7$.
- 根据观察到的$S=s$，`agent`在策略函函数的结果中随机抽样决定其行为。

`reward R`：`agent`做出一个动作`action`后给出一个奖励`reward`，这个奖励通常由我们定义，强化学习的目标即获得的奖励尽可能高，奖励定义的好坏非常影响强化学习的结果。

`state transition`：状态转换，当前状态下，`agent`采取一个行为，得到了一个新的状态。这个过程就称为状态转移，这个过程可以是固定的，也可以是随机的。
- 状态转移函数（条件概率密度函数）：$p(s^{'} | s, a) = \mathbb{P} (S^{'} = s^{'} | S = s, A = a)$

**Environment** ---(State $S_t$)---> `Agent` ---(Action $a_t$)---> **Environment** ---(State $S_{t+1}$, Reward $r_t$)---> `Agent` ---……

# Randomness in Reinforcement Learning

强化学习中的随机性
- `agent`采取的动作，动作是根据`policy`函数随机抽样得到;
- `state transition`，`agent`采取某动作后，可能得到多种状态，由系统随机抽样来选取最终状态。

例如游戏中：观察一帧（`state` $s_1$）
- 采取行为`action` $a_1$（前进、后退、跳跃……）；
- 得到新的一帧（`state` $s_2$）以及奖励`reward` $r_1$；
- 采取行为`action` $a_2$；
- ……

最终得到`(state, action, reward) trajectory`:
$s_1, a_1, r_1, s_2, a_2, r_2, ..., s_T, a_T, r_T$

# Rewards and Returns

`Reture`：回报（`cumulative future reward`未来累积奖励）
- $U_t = R_t + R_{t+1} + R_{t+2} + R_{t+3} + ...$

$R_t$和$R_{t+1}$同样重要吗？（今天的100元和明年的100对比？今天的80元和明年的100元对比？）

`Discounted return`：（`cumulative discounted future reward`）
- $\gamma$：折扣率（超参数，对强化学习有一定影响）；
- $U_t = R_t + \gamma R_{t+1} + \gamma ^ 2 R_{t+2} + \gamma ^ 3 R_{t+3} + ...$

对于时间`t`，回报$U_t$是随机的：
- `action`可能是随机的：$\mathbb{P} [A=a | S=s] = \pi(a|s)$.
- `state`可能是随机的：$\mathbb{P} [S^{'} = s^{'} | S=s, A=a] = p(s^{'}|s, a)$.
- 对于任意$i \le t$，奖励$R_i$取决于$S_i$和$A_i$；
- 于是给定$S_t$，其回报$U_t$取决于一系列随机变量：$A_t, A_{t+1}, A_{t+2}, ...$和$S_{t+1}, S_{t+2}, ...$。

# Value Function

$U_t = R_t + \gamma R_{t+1} + \gamma ^ 2 R_{t+2} + \gamma ^ 3 R_{t+3} + ...$

`Action-Value Function`（描述采取的动作好坏）：如果采用`policy`函数$\pi$，$Q_\pi$说明`agent`正处在状态`s`时做出动作`a`是否明智，即给动作`a`打分 ，$Q_\pi(s_t, a_t) = \mathbb E [U_t | S_t = s_t, A_t = a_t]$.
- 将$U_t$当作未来所有动作$a$和所有状态$s$的函数，未来动作$a$和状态$s$都具有随机性：
    - $\mathbb{P} [A=a | S=s] = \pi(a|s)$.
    - $\mathbb{P} [S^{'} = s^{'} | S=s, A=a] = p(s^{'}|s, a)$.

`Optiomal action-value function`：最优动作价值函数 $Q^*(s_t, a_t) = \max_\pi Q_\pi (s_t, a_t)$.

`State-value function`（描述当前局势好坏）与`policy`函数和当前状态$s$有关，与动作无关，也可以通过$E_S[V_\pi(S)]$ 评判`policy`函数$\pi$的好坏（$\pi$越好，$V_\pi$的平均值越大）：$V_\pi (s_t) = \mathbb E_A[Q_\pi (s_t, A)] = \sum_a \pi (a | s_t) \cdot Q_\pi (s_t, a) = \int \pi(a | s_t) \cdot Q_\pi (s_t, a) da$.

# How does AI control the agent?

- 假设我们有一个好的策略函数`policy` $\pi(a|s)$
    - 根据观察到的状态`state` $s_t$
    - 随机抽样行为 $a_t \sim \pi(\cdot | S_t)$

- 假设我们知道最优行为价值函数$Q^*(s, a)$
    - 根据观察到的状态`state` $s_t$
    - 选择让价值最大的行为 $a_t = argmax_a Q^*(s_t, a)$

学习到两种函数上任意一个即可。
