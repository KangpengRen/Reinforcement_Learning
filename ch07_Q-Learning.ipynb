# Q-Learning

## Sarsa VS Q-Learning

- `Sarsa`算法用于训练动作价值函数$Q_\pi(s, a)$；
- `TD target`：$y_t = r_t + \gamma \cdot Q_\pi(s_{t+1}, a_{t+1})$；
- 价值网络为动作价值函数$Q_\pi(s, a)$的近似。

- `Q-Learning`用于训练最优价值函数$Q^*(s, a)$；
- `TD target`：$y_t = r_t + \gamma \cdot max_a Q^*(s_{t+1}, a)$；
- `DQN`是对最优动作价值$Q^*$的近似。

## Derive TD Target

上一章中得到等式：
$$
Q_\pi(s_t, a_t) = \mathbb E [R_t + \gamma \cdot Q_\pi(S_{t+1}, A_{t+1})]
$$
记最优策略为$\pi^*$，得到：
$$
Q_{\pi^*}(s_t, a_t) = \mathbb E [R_t + \gamma \cdot Q_{\pi^*}(S_{t+1}, A_{t+1})]
$$
通常记$Q_{\pi^*}$为$Q^*$，于是：
$$
Q^*(s_t, a_t) = \mathbb E [R_t + \gamma \cdot Q^*(S_{t+1}, A_{t+1})]
$$
处理期望中的$Q^*$：
$$
\begin{align}
A_{t+1} = argmax_a Q^*(S_{t+1}, a) \\
Q^*(S_{t+1}, A_{t+1}) = max_a Q^*(S_{t+1}, a) \\
Q^*(s_t, a_t) = \mathbb E [R_t + \gamma \cdot max_a Q^*(S_{t+1}, a)]
\end{align}
$$
对期望做蒙特卡洛近似，得到`TD target`：
- $R_t \approx r_t$
- $max_a Q^*(S_{t+1}, a) \approx max_a Q^*(s_{t+1}, a)$
- $y_t = \mathbb E[r_t + max_a Q^*(s_{t+1}, a)] \approx\mathbb E [R_t + \gamma \cdot max_a Q^*(S_{t+1}, a)]$

## Q-Learning: Tabular Version

|             | Action $a_1$ | Action $a_2$ | Action $a_3$ | Action $a_4$ | ... |
|-------------|--------------|--------------|--------------|--------------|-----|
| State $s_1$ |              |              |              |              |     |
| State $s_2$ |              |              |              |              |     |
| State $s_3$ |              |              |              |              |     |
| ...         |              |              |              |              |     |

$max_a Q^*(s_{t+1}, a)$即第$t+1$行最大值。
- 观测到一个四元组`transition`：$(s_t, a_t, r_t, s_{t+1})$；
- 计算`TD target`：$y_t = r_t + \gamma \cdot max_a Q^*(s_{t+1}, a)$；
- 计算`TD error`：$\delta_t = Q^*(s_t, a_t) - y_t$；
- 更新表格：$Q^*(s_t, a_t) \leftarrow Q^*(s_t, a_t) - \alpha \cdot \delta_t$.

## Q-Learning: DQN Version

- `DQN`$Q(s, a; w)$是对最有动作价值函数$Q^*(s, a)$的近似。输入为当前状态，输出为对所有行为的评分；
- `DQN`可以用来控制`agent`：$a_t = argmax_a Q(s_t, a; w)$；
- 观察一个`transition`：$(s_t, a_t, r_t, s_{t+1})$；
- 计算`TD target`：$y_t = r_t + \gamma \cdot max_a Q^*(s_{t+1}, a; w)$；
- 计算`TD error`：$\delta_t = Q^*(s_t, a_t; w) - y_t$；
- 更新：$w \leftarrow w - \alpha \cdot \delta_t \cdot \frac {\partial Q(s_t, a_t; w)}{\partial w}$.
