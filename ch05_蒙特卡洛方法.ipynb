# Monte Carlo Algorithms

## Calculating Pi

通过蒙特卡洛方法计算$\pi$，假设我们有一个随机数生成器，随机生成$(-1, 1)$的实数，每次生成两个，一个作为$x$，一个作为$y$，每次生成的点都落在一个正方形中。

由于$x$和$y$都在$(-1, 1)$之间均匀分布，故正方形中所有的点都有相同的概率密度。

正方形中包含一个半径为1的圆，圆心在远点，生成的点可能落在圆中，也可能落在圆外。

- 正方形的面积：$A_1 = 2 ^ 2 = 4$
- 圆的面积：$A_2 = \pi r ^ 2 = \pi$

点落在圆中的概率：$P = \frac {A_2} {A_1} = \frac {\pi} {4}$.

在正方形中均匀抽样$n$个点，落在圆中的数量的期望为$Pn = \frac {\pi n} {4}$.

观测到$m$个点落在圆中，当$n$足够大时，$m \approx \frac {\pi n} 4$，于是$\pi \approx \frac {4m} n$.

```python
import random

n = 10000000
m = 0

for i in range(n):
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
    if x ** 2 + y ** 2 < 1:
        m += 1

pi = 4 * m / n
print(pi)
```

## Estimate of Expectation

- 定义$X$为$d$维随机变量；
- 函数$p(x)$为概率密度函数，描述变量$X$在某个取值点附近的可能性；
    - $\int_{\mathbb R d} p(x) dx = 1$.
- 期望：$\mathbb E_{X \sim p}[f(X)] = \int_{\mathbb R d} f(X) \cdot p(x) dx$.

通过蒙特卡洛方法近似期望：

- 根据概率密度函数$p(x)$抽样$x_1, x_2, ..., x_n$；
- 计算$Q_n = \frac 1 n \sum_{i=1}^n f(x_i)$；
- 返回$Q_n$作为对期望$\mathbb E_{X \sim p}[f(X)]$的估计。
