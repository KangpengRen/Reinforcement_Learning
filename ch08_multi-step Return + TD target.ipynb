# Multi-Step Return

$$
U_t = R_t + \gamma \cdot U_{t+1} = R_t + \gamma \cdot (R_{t+1} + \gamma \cdot U_{t=2}) = ...
$$
即：
$$
\begin{align}
U_t = R_t + \gamma \cdot R_{t+1} + \gamma^2 \cdot U_{t=2}. \\
U_t = R_t + \gamma \cdot R_{t+1} + \gamma^2 \cdot R_{t+2} + \gamma^3 \cdot U_{t+3}. \\
... \\
U_T = \sum_{i=0}^{m-1} \gamma^i \cdot R_{t+i} + \gamma^m \cdot U_{t+m}.
\end{align}
$$

将`m-step TD target`用于`Sarsa`：
$$
y_t = \sum_{i=0}^{m-1} \gamma^i \cdot r_{t+i} + \gamma^m \cdot Q_\pi(s_{t+m}, a_{t+m}).
$$

将`m-step TD target`用于`Q-Learning`：
$$
y_t = \sum_{i=0}^{m-1} \gamma^i \cdot r_{t+i} + \gamma^m \cdot max_a Q^*(s_{t+m}, a).
$$

- `One-step TD target`只用到一个观测到的奖励：$r_t$；
- `m-step TD target`用到了$m$个观测到的奖励：$r_t, r_{t+1}, ..., r_{t+m-1}$;
- 多步`m-step TD target`通常比`One-step TD target`更加接近真相，偏差更小，训练效果更好。
